```python
import pandas as pd

try:
    df_dec = pd.read_csv('Budget_FY2024_25 - Dec\'24.csv')
    df_feb = pd.read_csv('Budget_FY2024_25 - Feb\'25.csv')
    df_jan = pd.read_csv('Budget_FY2024_25 - Jan\'25.csv')
    df_mar = pd.read_csv('Budget_FY2024_25 - Mar\'25.csv')

    display(df_dec.head())
    display(df_feb.head())
    display(df_jan.head())
    display(df_mar.head())

except FileNotFoundError:
    print("One or more CSV files not found.")
except pd.errors.ParserError:
    print("Error parsing one or more CSV files. Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import pandas as pd
merged_df = pd.merge(pd.merge(pd.merge(df_dec, df_feb, on=['AccountId', 'AccountName', 'Owner'], how='outer'),
                             df_jan, on=['AccountId', 'AccountName', 'Owner'], how='outer'),
                    df_mar, on=['AccountId', 'AccountName', 'Owner'], how='outer')

monthly_columns = [col for col in merged_df.columns if col.endswith('2025') or col.endswith('2024')]
merged_df[monthly_columns] = merged_df[monthly_columns].fillna(0)

display(merged_df.head())

filtered_df = merged_df[merged_df['AccountId'] == 463470958933]
display(filtered_df)
```

```python
# Calculate the sum of each month, ignoring NaNs
dec_sum = merged_df['Dec 2024'].sum(skipna=True)
jan_sum = merged_df['Jan 2025'].sum(skipna=True)
feb_sum = merged_df['Feb 2025'].sum(skipna=True)
mar_sum = merged_df['Mar 2025'].sum(skipna=True)

# Calculate the average of the four monthly sums
average = (dec_sum + jan_sum + feb_sum + mar_sum) / 4

# Create a new row for the totals
total_row = pd.DataFrame({
    'AccountId': [999999],
    'AccountName': ['Total'],
    'Owner': ['Total'],
    'Mar 2025': [mar_sum],
    'Feb 2025': [feb_sum],
    'Jan 2025': [jan_sum],
    'Dec 2024': [dec_sum]
})

# Append the total row to the merged dataframe
merged_df = pd.concat([merged_df, total_row], ignore_index=True)

# Display the updated dataframe
display(merged_df)
```

```python
# Calculate the average of the four months for each account, excluding the total row
merged_df['Average'] = merged_df[['Dec 2024', 'Jan 2025', 'Feb 2025', 'Mar 2025']].mean(axis=1, skipna=True)

# Create new columns for the remaining months and populate with the average
month_year_cols = [
    "Apr 2024", "May 2024", "Jun 2024", "Jul 2024", "Aug 2024", "Sep 2024", "Oct 2024", "Nov 2024","Apr 2025", "May 2025", "Jun 2025", "Jul 2025",
    "Aug 2025", "Sep 2025", "Oct 2025", "Nov 2025", "Dec 2025", "Jan 2026",
    "Feb 2026", "Mar 2026"
]

for col in month_year_cols:
    merged_df[col] = merged_df['Average']

new_column_order = [
    "AccountId", "AccountName", "Owner","Average", "Mar 2025", "Feb 2025", "Jan 2025", "Dec 2024","Nov 2024", "Oct 2024", "Sep 2024", "Aug 2024",
    "Jul 2024", "Jun 2024", "May 2024", "Apr 2024", "Apr 2025", "May 2025",
    "Jun 2025", "Jul 2025", "Aug 2025", "Sep 2025", "Oct 2025", "Nov 2025",
    "Dec 2025", "Jan 2026", "Feb 2026", "Mar 2026"
]

merged_df = merged_df.reindex(columns=new_column_order)
display(merged_df)
```

```python
merged_df.to_csv('merged_budget_data.csv', index=False)
```
